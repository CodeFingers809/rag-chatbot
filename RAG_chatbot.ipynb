```python
#loading documents
from langchain_community.document_loaders import TextLoader, PyPDFLoader
import os

my_dir = "/content/books"
docs = []
for f in os.listdir(my_dir):
    if not f.endswith(".txt"): continue
    print(f"\n{f}")
    loadertxt = TextLoader(os.path.join(my_dir, f))
    doc = loadertxt.load()
    print(type(doc), len(doc))
    print(type(doc[0]))
    print(doc[0].metadata)
    print(len(doc[0].page_content))
    docs.extend(doc)
print(len(docs))
```

```python
#chunking
!pip install -qU langchain-text-splitters
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

splitter = RecursiveCharacterTextSplitter(chunk_size = 1000, chunk_overlap=100)
chunks = splitter.split_documents(docs)

print(f"{len(doc)} documents have been split into {len(chunks)} chunks.")
print(f"Each chunk is of type {type(chunks[0])}")
print(f"\nSample chuck page content:\n{chunks[0].page_content}")
```

```python
!pip install -qU sentence_transformers
```

```python
# Embeddings and Vectorstore
from langchain_community.embeddings.sentence_transformer import SentenceTransformerEmbeddings

embeddings = SentenceTransformerEmbeddings(model_name = "all-MiniLM-L6-v2")
```

```python
!pip install -qU langchain_chroma
```

```python
from langchain_chroma import Chroma

db = Chroma.from_documents(chunks, embeddings)
persist_directory = os.path.join(my_dir, "chroma_db")
vectordb = Chroma.from_documents(
    documents=chunks,
    embedding=embeddings,
    persist_directory=persist_directory
)
```

```python
# Retreival
query = "Summarize what does it mean to be earnest?"
relevant_docs = vectordb.max_marginal_relevance_search(query)
print(type(relevant_docs), len(relevant_docs))
print(type(relevant_docs[0]))
```

```python
for d in relevant_docs:
    print(f"--------------------------{len(d.page_content)}----------------------\n{d.page_content}")
```

```python
from langchain_groq import ChatGroq
from langchain_core.prompts import ChatPromptTemplate

llm = ChatGroq(
    model = "llama3-70b-8192",
    temperature=0,
    api_key=GROQ_API_KEY
)

system = "You are a helpful assistant. Answer the user's query based on the given context. DO NOT MAKE ANSWERS UP."
human = """Answer the following question based on the given context:
{question}

Context: {context}"""
prompt = ChatPromptTemplate.from_messages([
    ("system", system),
    ("human", human)
])

chain = ChatPromptTemplate.from_messages([
    ("system", system),
    ("human", human)
])

chain = prompt | llm
```

```python
#relevant docs is a list of Document Objects
context = "\n".join([d.page_content for d in relevant_docs])

response = chain.invoke({"question": query, "context": context})
print(response.content)
```
